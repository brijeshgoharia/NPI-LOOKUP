```python
#Import required libraries
import requests
import pandas as pd

#Import the data from the excel file
df1 = pd.read_excel("/Users/brijeshgoharia/Desktop/KMK CONSULTING INC/NPI_PROJECT/HCP_list1.xlsx")
df1 = df1[["First", "Name", "City", "State", "Zip Code"]]
```

```python
df1.head(5)
```

```python
#empty list to store the response
response_dfs = []

#iterating through each row in the dataframe
for i, row in df1.iterrows():

#preparing data to feed in the URL link    
    city =row["City"]
    city = city.split(" ")
    s = "+"
    city = s.join(city)
    
        
    state =row["State"]
    state = state.split(" ")
    s = "+"
    state = s.join(state)
    
    first =row["First"]
    first = first.split(" ")
    s = "+"
    first = s.join(first)
    
    last =row["Name"]
    last = last.split(" ")
    s = "+"
    last = s.join(last)+str("*")
    
    query = """https://npiregistry.cms.hhs.gov/api/?version=2.1&state={}&first_name={}&last_name={}""".format(state,first,last)
    print(query)
    rez1 = requests.get(query)
    rez1_df = pd.DataFrame(rez1.json())
    
    
#Neglecting those data not found     
    if rez1_df.empty:
        print("Empty")
        next
    else:
        rez2=rez1_df['results'][0]
        rez1_norm_df = pd.json_normalize(rez2)
        rez3=rez1_norm_df['addresses']
        rez2_norm_df = pd.json_normalize(rez3[0])
        data1=rez1_norm_df[['number','basic.first_name','basic.last_name',]]
        data1['state']=rez2_norm_df['state'].iloc[0]
        response_dfs.append(data1)
        #rez3 = pd.DataFrame() 
```

```python
response_dfs
```

```python
#Merging the response data with the input
df2 = pd.concat(response_dfs, axis=0)
df3=df2.rename(columns={"number": "NPI", "basic.first_name": "First", "basic.last_name": "Name", "state": "State"})


df3['First'] = df3['First'].str.lower()
df3['Name'] = df3['Name'].str.lower()

df1['First'] = df1['First'].str.lower()
df1['Name'] = df1['Name'].str.lower()


df4 = pd.merge(df1, df3, on=["First", "Name","State"], indicator=True, how="left")
df4
```
